### Image to Pencil Sketch

```python
# Importing the libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'anime1.jpeg'
image = cv2.imread(image_path)

# Converting BGR image to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Displaying the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
def pencil_sketch(image_path):
  # Loading the image
  image = cv2.imread(image_path)

  # Converting the image to grayscale
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Inverting the grayscale image
  inverted_gray_image = 255 - gray_image

  # Applying a blur filter to the inverted image
  blurred_image = cv2.GaussianBlur(inverted_gray_image, (21, 21), 0)

  # Combining the grayscale image and the blurred image
  pencil_sketch_image = cv2.divide(gray_image, 255 - blurred_image, scale=256.0)

  return pencil_sketch_image

```

```python
# Getting the image
image_path = "anime1.jpeg"

# Creating the pencil sketch
sketch = pencil_sketch(image_path)

# Displaying the original image and pencil sketch image
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(sketch, cmap='gray')
axes[1].set_title('Pencil sketch')
axes[1].axis('off')

plt.show()
```
